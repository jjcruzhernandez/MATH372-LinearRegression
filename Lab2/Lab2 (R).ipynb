# Lab 2: SLR with Bike Share Data (R)

```r
data <- read.csv(file = "bike_share.csv") 
countData <- data$count
tempData <- data$temp
humidityData <- data$humidity
windspeedData <- data$windspeed
```

```r
scatterPlot <- function(x, y, model, level, xlab){
    plot(x, y, ylab = "# Bike Rentals", xlab = xlab)
    abline(reg = model, lwd = 2, col = "red")
    conf <- data.frame(x, predict(model, interval = "confidence", level = level))
    conf <- conf[order(conf[,1]),]
    lines(conf[,1], conf[,3], col = "blue", lwd = 2, lty = 2)
    lines(conf[,1], conf[,4], col = "blue", lwd = 2, lty = 2)
    pred <- data.frame(x, predict(model, interval = "prediction", level = level))
    pred <- pred[order(pred[,1]),]
    lines(pred[,1], pred[,3], col = "purple", lwd = 2, lty = 2)
    lines(pred[,1], pred[,4], col = "purple", lwd = 2, lty = 2)
    legend("topleft", legend = c("Fitted Values", "95% CI", "95% PI"), col = c("red", "blue", "purple"), 
           lwd = 2, lty = c(1,2,2))
}
```
